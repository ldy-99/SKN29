```python
import pandas as pd
import numpy as np
df = pd.read_csv('data/messy_data.csv')
df['score'].interpolate(method='polynomial', order=3)
df.head()
```

```python
na_cols = df.isna().any()[df.isna().any()].index
na_cols

for col in na_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.shape
```

```python
def IQR_detect(data,column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - IQR*1.5
    upper = Q3 + IQR*1.5

    outlier_mask = (data[column] > upper) | (data[column] < lower)
    return outlier_mask , lower , upper , Q1 , Q3 , IQR
```

```python
numeric = df.describe().columns[1:]

for c in numeric:
    outlier_mask , lower , upper , Q1 , Q3 , IQR = IQR_detect(df,c)
    print(c)
    print(outlier_mask.sum())
```

```python
from scipy import stats
stats.zscore(df['age'])
def detect_outlier_zscore(data,column,threshold=3):
    z_score = abs(stats.zscore(data[column].dropna()))
    return ~(z_score > threshold)
```

```python
df.columns
```

```python
for col in numeric :
    before = len(df)
    zscore_index = detect_outlier_zscore(df,col)
    df = df[zscore_index]
    removed = before - len(df)
    print(col,removed)
```
