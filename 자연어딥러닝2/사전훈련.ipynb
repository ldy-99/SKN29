```python
from gensim.models import Word2Vec
```

```python
sentences = [
    ['I' , 'love' , 'natural' , 'language' , 'processing'] ,
    ['natural' , 'language' , 'processing' , 'is' , 'fun'] ,
    ['I' , 'enjoy' , 'learning' , 'machine' , 'learning']
]

w2v_model = Word2Vec(sentences, vector_size=10 , window=2 , min_count=1)
print(f"language 벡터 : {w2v_model.wv['language']}")
print(f"language와 natural의 유사도 : {w2v_model.wv.similarity('language', 'natural')}")
```

```python
# 글로벌 동시 등장 확률 (co-occurrence) king - man + woman = queen
import gensim.downloader as api
print('Glove 모델 로딩중...')
glove_model = api.load("glove-wiki-gigaword-50")
print('Glove 모델 로딩 완료!')
```

```python
print(f"king , queen 유사도 : {glove_model.similarity('king', 'queen')}")
```

```python
from numpy import positive
from torch import ne

res = glove_model.most_similar(positive=['king','woman'], negative=['man'], topn=3)
print(f"king-man+woman 에 가장 가까운 단어: {res}")
```

```python
from gensim.models import FastText
ft_model = FastText(sentences, vector_size=10 , window=2 , min_count=1 ,min_n=3 , max_n= 5 , epochs=1000)
print(ft_model.wv['language'])
print(ft_model.wv['langu'])
print(ft_model.wv.similarity('language', 'langu'))
```
